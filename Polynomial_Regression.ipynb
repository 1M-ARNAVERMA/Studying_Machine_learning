```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data_set = pd.read_csv('RBC in HPR.csv')
x = data_set.iloc[ : , :-1].values
y = data_set.iloc[ : , -1 ].values
```

### Training the Linear regression model on dataset

```python
from sklearn.linear_model import LinearRegression
linear_reg = LinearRegression()
linear_reg.fit(x, y) 
```

### Training the polynomial regreesion model

```python
from sklearn.preprocessing import PolynomialFeatures
polynomial_reg = PolynomialFeatures(degree = 4)  # Degree of the polynomial is 4
x_poly = polynomial_reg.fit_transform(x)

linear_reg_2 = LinearRegression()
linear_reg_2.fit(x_poly , y)
```

### Visualize Linear Regression

```python
plt.scatter(x,y,color= 'red')
plt.plot(x, linear_reg.predict(x),color = 'blue')
plt.title('RBC = Linear Regression')
plt.xlabel('Age')
plt.ylabel('RBC')
plt.show()

# This model FAILING as we can see
```

### Visualizing polynomial regression

```python
plt.scatter(x,y,color= 'red')
plt.plot(x, linear_reg_2.predict(x_poly),color = 'blue')
plt.title('RBC = Polynomial Regression')
plt.xlabel('Age')
plt.ylabel('RBC')
plt.show()
```

### Predicting a new result with Linear Regression 

```python
# Normal amount of RBC in a child of year 6.5
linear_reg.predict([[6.5]])

# Answer is wrong here
```

### Predicting a new result with Polynomial Regression 

```python
linear_reg_2.predict(polynomial_reg.fit_transform([[6.5]]))
```
