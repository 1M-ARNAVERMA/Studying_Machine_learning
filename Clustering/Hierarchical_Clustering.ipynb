```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
```

```python
dataset = pd.read_csv('Amazon.com Clusturing Model.csv')
x = dataset.iloc[: , [2,4]].values
```

```python
print(x)
```

### Optimal number of cluseters

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(x,method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Age')
plt.ylabel('Euclidean Distance')
```

### Hierarchical CLustering model training

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=6, metric='euclidean', linkage='ward')
yhc = hc.fit_predict(x)
```

```python
plt.scatter(x[yhc == 0 ,0],x[yhc == 0,1], s = 100, c = 'magenta', label = 'Cluster 1')
plt.scatter(x[yhc == 1 ,0],x[yhc == 1,1], s = 100, c = 'red', label = 'Cluster 2')
plt.scatter(x[yhc == 2 ,0],x[yhc == 2,1], s = 100, c = 'blue', label = 'Cluster 3')
plt.scatter(x[yhc == 3 ,0],x[yhc == 3,1], s = 100, c = 'yellow', label = 'Cluster 4')
plt.scatter(x[yhc == 4 ,0],x[yhc == 4,1], s = 100, c = 'green', label = 'Cluster 5')
plt.scatter(x[yhc == 5 ,0],x[yhc == 5,1], s = 100, c = 'black', label = 'Cluster 6')

plt.title('Amazon users')
plt.xlabel('Age')
plt.ylabel('Rating')
plt.legend
plt.show()
```
