```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data_set = pd.read_csv('Salary_DTR.csv')
x = data_set.iloc[ : , 1:-1].values
y = data_set.iloc[ : , -1 ].values
```

```python
print(x)
print(y)
```

### Training the Dataset

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 12 , random_state = 42)  # How many time do we need to do it
regressor.fit(x,y)
```

```python
regressor.predict([[5.5]])
```

```python
x_grid = np.arange(min(x),max(x),0.01)
x_grid = x_grid.reshape(len(x_grid),1)

plt.scatter(x,y,color = 'red')
plt.plot(x_grid, regressor.predict(x_grid) , color = 'blue')
plt.title('Decision Tree Regression')
plt.xlabel('Grade')
plt.ylabel('Salary')
plt.show()

# No. of steps are increased here as compared to decision tree
```
